```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Create the dataset
data = {
    'area': [2600, 3000, 3200, 3600, 4000],
    'price': [550000, 565000, 610000, 680000, 725000]
}

# Convert the dataset to a DataFrame
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv('homeprices.csv', index=False)

# Load the dataset
homeprices = pd.read_csv('homeprices.csv')

# Prepare the data
X = homeprices[['area']]
Y = homeprices['price']

# Train the linear regression model
model = LinearRegression()
model.fit(X, Y)

# Predict prices for the given areas
predicted_prices = model.predict(X)

# Display the results
results = pd.DataFrame({
    'area': homeprices['area'],
    'actual_price': homeprices['price'],
    'predicted_price': predicted_prices
})

print(results)

# If needed, save the predictions to a new CSV file
results.to_csv('predicted_homeprices.csv', index=False)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np

# Generate a sample dataset and save it as income.csv
data = {
    'Feature1': [15, 20, 18, 30, 40, 42, 45, 70, 65, 78],
    'Feature2': [39, 36, 30, 80, 87, 91, 88, 160, 155, 170]
}
df = pd.DataFrame(data)
df.to_csv('income.csv', index=False)
print("Sample dataset 'income.csv' has been created.")

# Load the dataset
data = pd.read_csv('income.csv')

# Check the structure of the dataset
print("Dataset preview:")
print(data.head())

# Extract relevant features for clustering
X = data[['Feature1', 'Feature2']]

# Define the K-Means model with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model to the data
kmeans.fit(X)

# Assign clusters to the data
data['Cluster'] = kmeans.labels_

# Calculate new centroids based on the mean values of the coordinates of each cluster
new_centroids = data.groupby('Cluster').mean()[['Feature1', 'Feature2']]

# Display the new centroids
print("\nNew centroid values:")
print(new_centroids)

# Save the clustered data to a new CSV file
data.to_csv('clustered_income.csv', index=False)
print("\nClustered data saved to 'clustered_income.csv'.")
```

```python
import pandas as pd
from sklearn.cluster import KMeans

# Define the dataset
data = {
    "Objects": ["OB-1", "OB-2", "OB-3", "OB-4", "OB-5", "OB-6", "OB-7", "OB-8"],
    "X": [1, 1, 1, 2, 1, 2, 1, 2],
    "Y": [4, 2, 4, 1, 1, 4, 1, 1],
    "Z": [1, 2, 2, 2, 1, 2, 2, 1]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Extract the feature columns
X = df[["X", "Y", "Z"]]

# Apply K-Means clustering with K=2
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)

# Add cluster labels to the DataFrame
df["Cluster"] = kmeans.labels_

# Calculate the new centroids
centroids = pd.DataFrame(kmeans.cluster_centers_, columns=["X", "Y", "Z"])

# Print results
print("Clustered Data:")
print(df)
print("\nNew Centroids:")
print(centroids)

# Save clustered data to a CSV file
df.to_csv("clustered_objects.csv", index=False)
print("\nClustered data saved to 'clustered_objects.csv'.")
```

```python
# Import necessary libraries
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Sample dataset in the form of a list of transactions
data = {
    'Bread': [1, 1, 1, 0, 0],
    'Milk': [1, 1, 0, 1, 1],
    'Cheese': [1, 0, 0, 1, 1],
    'Butter': [0, 1, 1, 1, 0],
    'Jam': [0, 0, 1, 0, 1]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert the DataFrame values to boolean (True/False) for better performance
df = df.astype(bool)

# Generate frequent itemsets using Apriori with a support threshold of at least 5%
frequent_itemsets = apriori(df, min_support=0.05, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# Output the results
print("Frequent Itemsets:")
print(frequent_itemsets)

print("\nAssociation Rules:")
print(rules)

# Save outputs to CSV files (optional)
frequent_itemsets.to_csv("frequent_itemsets.csv", index=False)
rules.to_csv("association_rules.csv", index=False)

print("\nOutputs saved to 'frequent_itemsets.csv' and 'association_rules.csv'.")
```

```python
# Import necessary libraries
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Sample dataset in the form of a list of transactions
data = {
    'Bread': [1, 1, 1, 0, 0],
    'Milk': [1, 1, 0, 1, 1],
    'Cheese': [1, 0, 0, 1, 1],
    'Butter': [0, 1, 1, 1, 0],
    'Jam': [0, 0, 1, 0, 1]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Convert the DataFrame values to boolean (True/False) for better performance
df = df.astype(bool)

# Generate frequent itemsets using Apriori with a support threshold of at least 5%
frequent_itemsets = apriori(df, min_support=0.05, use_colnames=True)

# Generate association rules without specifying the 'num_itemsets' argument
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

# Output the results
print("Frequent Itemsets:")
print(frequent_itemsets)

print("\nAssociation Rules:")
print(rules)

# Save outputs to CSV files (optional)
frequent_itemsets.to_csv("frequent_itemsets.csv", index=False)
rules.to_csv("association_rules.csv", index=False)

print("\nOutputs saved to 'frequent_itemsets.csv' and 'association_rules.csv'.")
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sample dataset with objects and their X, Y, Z coordinates
data = {
    'Objects': ['OB-1', 'OB-2', 'OB-3', 'OB-4', 'OB-5', 'OB-6', 'OB-7', 'OB-8'],
    'X': [1, 1, 1, 2, 1, 2, 1, 2],
    'Y': [4, 2, 4, 1, 1, 4, 1, 1],
    'Z': [1, 2, 2, 2, 1, 2, 2, 1]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Extract X, Y, Z coordinates for clustering
coordinates = df[['X', 'Y', 'Z']]

# Apply K-Means clustering with K=3
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(coordinates)

# Output the DataFrame with the clusters
print("Clustered DataFrame:")
print(df)

# 3D Plot to visualize the clustering result
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot the points with color corresponding to their cluster
scatter = ax.scatter(df['X'], df['Y'], df['Z'], c=df['Cluster'], cmap='viridis')

# Add labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('K-Means Clustering (K=3)')

# Show color bar for clusters
fig.colorbar(scatter)

# Show the plot
plt.show()
```
